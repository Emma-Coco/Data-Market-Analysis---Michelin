```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le fichier
fichier = "../Clean_data/Requêtes_all.xlsx"
requetes_df = pd.read_excel(fichier)
pd.set_option('display.max_columns', None)  # Pour voir toutes les colonnes
pd.set_option('display.width', 1000)        # Pour éviter les retours à la ligne
pd.set_option('display.precision', 4)       # Pour les nombres décimaux

# Nettoyage
requetes_df.rename(columns={"Requêtes_les_plus_fréquentes": "Requete"}, inplace=True)

requetes_df = requetes_df[requetes_df['Source'] != 'Global']
# Renommer la colonne pour simplifier
print(requetes_df.head(10))
```

```python
# 1. Top 10 requêtes par nombre de clics
plt.figure(figsize=(12, 6))
sns.barplot(data=requetes_df.sort_values("Clics", ascending=False).head(10), y="Requete", x="Clics", palette="Blues_d")
plt.title("Top 10 des requêtes par clics")
plt.xlabel("Clics")
plt.ylabel("Requête")
plt.tight_layout()
plt.show()

# 2. Top 10 requêtes par impressions
plt.figure(figsize=(12, 6))
sns.barplot(data=requetes_df.sort_values("Impressions", ascending=False).head(10), y="Requete", x="Impressions", palette="Oranges")
plt.title("Top 10 des requêtes par impressions")
plt.xlabel("Impressions")
plt.ylabel("Requête")
plt.tight_layout()
plt.show()

# 3. Top 10 requêtes par CTR
plt.figure(figsize=(12, 6))
sns.barplot(data=requetes_df.sort_values("CTR", ascending=False).head(10), y="Requete", x="CTR", palette="Greens")
plt.title("Top 10 des requêtes par taux de clics (CTR)")
plt.xlabel("CTR")
plt.ylabel("Requête")
plt.tight_layout()
plt.show()
```

```python
# 4. Scatter plot : Position vs CTR (taille = clics)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=requetes_df, x="Position", y="CTR", size="Clics", hue="Source", alpha=0.7, sizes=(20, 300))
plt.title("CTR vs Position moyenne des requêtes")
plt.xlabel("Position moyenne")
plt.ylabel("CTR")
plt.tight_layout()
plt.show()
```

```python
# 5. Focus sur requêtes stratégiques : haut CTR + mauvaise position
top_strategique = requetes_df.sort_values(by=["CTR", "Position"], ascending=[False, True]).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top_strategique, x="CTR", y="Requete", palette="coolwarm")
plt.title("Requêtes stratégiques : fort CTR mais faible visibilité")
plt.xlabel("CTR")
plt.ylabel("Requête")
plt.tight_layout()
plt.show()

#needs fixinng
```
